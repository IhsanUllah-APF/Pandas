```python
# Method 2: Use groupby.
```

```python
df.groupby(['Country'])
# DataFrameGroupBy object created and stored in the memory location 0x00000204.... 
# What actually has happened, as Corey said, the rows have been grouped by countries. That is, the data in the dataframe
# has been grouped (divided) based on countries.
```

```python
df.groupby(['Country']).get_group('United States')
# Grab only those rows from the dataframe where the country is United States.
```

```python
df.groupby(['Country']).get_group('India')
# just by typing the country, we can obtain the observations for the desirable country.
# The same can be done by using filter, but it will require changing filter again and again.
```

```python
df.groupby(['Country']).get_group('India')['SocialMedia']
# Grab values from the above dataframe only for SocialMedia. 
```

```python
df.groupby(['Country']).get_group('India')['SocialMedia'].value_counts() 
```

```python
# A variant of groupby method.
```

```python
# Next method to obtain the same information as immediately above.
df.groupby(['Country'])['SocialMedia']
# df.groupby(['Country']) has already grouped/divided the data in dataframe based on countries. ['SocialMedia'] takes
# the column, SocialMedia, from same dataframe, already grouped on countries, and stores it as a SeriesGroupBy object at 
# memory location 0x00000204.... 
```

```python
df.groupby(['Country'])['SocialMedia'].value_counts()
# grab the value counts/ frequencies for Social Media from the seires already stored as an object above, which contains
# the data/values in SocialMedia column grouped on countries.
```

```python
df.groupby(['Country'])['SocialMedia'].value_counts().head(50)
```

```python
df.groupby(['Country'])['SocialMedia'].value_counts().loc['India']
# Countries are acting as an index above. So, with .loc, we can find the SocialMedia data for any country. 
```

```python
# What percent of respondents in each country know or use Python? See below.
```

```python
# First, find the total number of respondents in each country.
df['Country']
```

```python
df['Country'].value_counts()
```

```python
# Second, first total number of python users in each country.
df.groupby('Country')['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondents = df['Country'].value_counts()
```

```python
country_python_users = df.groupby('Country')['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_respondents
```

```python
country_python_users
```

```python
python_df = pd.concat([country_respondents, country_python_users]) # concated row wise and returned a series
```

```python
python_df
# Since concated row wise, so the number of rows have increased two times (179 * 2 =) 358
```

```python
python_df['Afghanistan']
```

```python
python_df['United States']
```

```python
python_df = pd.concat([country_respondents, country_python_users], axis = 'columns', sort = False)
#concated column wise and returned a dataframe. sort = False not needed.
```

```python
python_df
# Since concated column wise, therefore, the number of rows have remained same (179) and columns have increased from
# 1 previously to 2 now.
```

```python
python_df.rename(columns = {'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace = True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by = ['PctKnowsPython'], ascending = False) # can do inplace = True, or head() etc like what we can
# do with any dataframe.
```
